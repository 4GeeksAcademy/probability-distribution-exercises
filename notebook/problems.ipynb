## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Normal Distribution

```python
from scipy.stats import norm

# Define all the probability functions related to this distribution

def dnorm(x, loc=0, std = 1):
    """    Density of the normal distribution    """
    result = norm.pdf(x, loc = 0, scale = std)
    return result

def pnorm(x, loc=0, std = 1):
    """    Cumulative of the normal distribution    """
    result = norm.cdf(x = x, loc = 0, scale = std)
    return result

def percnorm(p=0.95, loc = 0, std = 1):
    """    Percentile function from the normal distribution    """
    result = norm.ppf(q = p, loc = 0, scale = std)
    return result

def qnorm(p, loc = 0, std = 1):
    """    Quantile function from the normal distribution    """
    result = norm.ppf(q = p, loc = 0, scale = std)
    return result

def snorm(p, loc = 0, std = 1):
    """    Survival function from the normal distribution    """
    result = norm.sf(q = p, loc = 0, scale = std)
    return result

def rnorm(n, loc = 0, std = 1):
    """    Generates random variables from the normal distribution    """
    result = norm.rvs(size = n, loc = 0, scale = std)

    return result
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

loc = 0
std = 1
data = np.arange(-10, 10, 0.01)

pmf = dnorm(data, loc = loc, std = std)
cdf = pnorm(data, loc = loc, std = std)
perc = percnorm(data, loc = loc, std = std)
ppf = qnorm(data, loc = loc, std = std)
#sf = snorm(data, loc = loc, std = std)

fig, axis = plt.subplots(4, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].plot(data, cdf, "g-")
axis[2].plot(data, perc, "b-")
axis[3].plot(data, ppf, "y-")
#axis[4].plot(data, sf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Uniform Distribution

```python
from scipy.stats import uniform

def duniform(x, loc=0, scale=1):
    """    Calculates the point estimate of the uniform distribution    """
    result = uniform.pdf(x, loc, scale)
    return result

def puniform(q, loc=0, scale=1):
    """    Cumulative of the uniform distribution    """
    result = uniform.cdf(q, loc, scale)
    return result

def percuniform(p=0.95, loc = 0, scale = 1):
    """    Percentile function from the normal distribution    """
    result = uniform.ppf(p, loc, scale)
    return result

def suniform(p, loc = 0, scale = 1):
    """    Survival function from the normal distribution    """
    result = uniform.sf(p, loc, scale)
    return result

def runiform(n, loc=0, scale=1):
    """    Random variables from the uniform distribution    """
    result = uniform.rvs(loc, scale, size = n)
    return result
```

```python
np.random.seed(42)

data = np.arange(-10, 10, 0.01)

pmf = duniform(data, 0, 1)
cdf = puniform(data, 0, 1)
ppf = percuniform(data, 0, 1)
sf = suniform(data, 0, 1)

fig, axis = plt.subplots(4, 1, figsize = (5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "g-")
axis[2].plot(data, ppf, "b-")
axis[3].plot(data, sf, "y-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
